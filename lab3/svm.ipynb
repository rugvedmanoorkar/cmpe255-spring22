```python
np.random.seed(0)

data = pd.read_csv("../input/fetal-prediction/fetal_health.csv")
data.head()
```

```python
target = ['Normal','Suspect', 'Pathological']
fetal_col = data.fetal_health.astype(int)
print(fetal_col)
data['fetal_labels'] = fetal_col.apply(lambda x: target[x-1])
data.head()
```

```python
data0 = data[data.fetal_health == 1.0]
data1 = data[data.fetal_health == 2.0]
data2 = data[data.fetal_health == 3.0]
data0.head()

```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['fetal_health','fetal_labels'], axis='columns')
y = data.fetal_health
X_scaled = sc.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn import svm
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
model_C = SVC(kernel='linear', C=1)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_C1 = SVC(kernel='linear',C=10)
model_C1.fit(X_train, y_train)
model_C1.score(X_test, y_test)
```

```python
model_C2 = SVC(kernel='linear', C=50)
model_C2.fit(X_train, y_train)
model_C2.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 8))

sns.scatterplot(x=X_train[:, 0], 
                y=X_train[:, 1], 
                hue=y_train, 
                s=8);

w = model_C2.coef_[0]          
b = model_C2.intercept_[0]     
x_points = np.linspace(-1, 1)    
y_points = -(w[0] / w[1]) * x_points - b / w[1]  

plt.plot(x_points, y_points, c='r');
```

```python
sns.scatterplot(x=X_train[:, 0], 
                y=X_train[:, 1], 
                hue=y_train, 
                s=8);
w = model_C.coef_[0]          
b = model_C.intercept_[0]     
x_points = np.linspace(-1, 1)    
y_points = -(w[0] / w[1]) * x_points - b / w[1]  

plt.plot(x_points, y_points, c='r');
```

```python
sns.scatterplot(x=X_train[:, 0], 
                y=X_train[:, 1], 
                hue=y_train, 
                s=8);
w = model_C1.coef_[0]          
b = model_C1.intercept_[0]     
x_points = np.linspace(-1, 1)    
y_points = -(w[0] / w[1]) * x_points - b / w[1]  

plt.plot(x_points, y_points, c='r');
```

```python
def plot(test, pred):
    X = np.array(test)
    y = np.array(pred)
    normal, suspect,pathological = X[np.argwhere(y==1.0)],X[np.argwhere(y==2.0)],X[np.argwhere(y==3.0)]
     
      
    plt.scatter([s[0][0] for s in normal],[s[0][1] for s in normal],s = 25,color = 'blue',edgecolor = 'k',marker = '^')
    plt.scatter([s[0][0] for s in suspect],[s[0][1] for s in suspect],s = 25,color = 'red',edgecolor = 'k',marker = 's')
    plt.scatter([s[0][0] for s in pathological],[s[0][1] for s in pathological], s = 25,color = 'yellow',edgecolor = 'k',marker = 's')
    plt.xlabel('X Test')
    plt.ylabel('Y Test')
    plt.legend(['Normal','Suspect','Pathological'])
```

```python
import sklearn.metrics as metrics
y_c = model_C.predict(X_test)
plot(X_test, y_c)
print("Accuracy for C = 1: ", metrics.accuracy_score(y_test, y_c))

```

```python
y_c10 = model_C1.predict(X_test)
plot(X_test, y_c10)
acc = metrics.accuracy_score(y_test, y_c10)
print("Accuracy for C = 10",  acc)
```

```python
y_c50 = model_C2.predict(X_test)
plot(X_test, y_c50)
acc = metrics.accuracy_score(y_test, y_c50)
print("Accuracy for C = 50: ", acc)
```
