```python
np.random.seed(0)

data = pd.read_csv("../input/fetal-prediction/fetal_health.csv")
data.head()
#data.describe()
```

```python
target = ['Normal','Suspect', 'Pathological']
fetal_col = data.fetal_health.astype(int)
print(fetal_col)
data['fetal_labels'] = fetal_col.apply(lambda x: target[x-1])
data.head()
```

```python
data0 = data[data.fetal_health == 1.0]
data1 = data[data.fetal_health == 2.0]
data2 = data[data.fetal_health == 3.0]
data0.head()

```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['fetal_health','fetal_labels'], axis='columns')
y = data.fetal_health
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)
```

```python
from sklearn import svm
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
model_C = SVC(kernel='linear', C=30)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_C1 = SVC(kernel='linear',C=20)
model_C1.fit(X_train, y_train)
model_C1.score(X_test, y_test)
```

```python
model_C2 = SVC(kernel='linear', C=1)
model_C2.fit(X_train, y_train)
model_C2.score(X_test, y_test)
```

```python
# get the separating hyperplane
w = model_C2.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-5, 5)
yy = a * xx - (model_C2.intercept_[0]) / w[1]
margin = 1 / np.sqrt(np.sum(model_C2.coef_ ** 2))
yy_down = yy - np.sqrt(1 + a ** 2) * margin
yy_up = yy + np.sqrt(1 + a ** 2) * margin
plt.figure(1, figsize=(4, 3))
plt.clf()
plt.plot(xx, yy, "k-")
plt.plot(xx, yy_down, "k-")
plt.plot(xx, yy_up, "k-")
plt.scatter(model_C2.support_vectors_[:, 0], model_C2.support_vectors_[:, 1], s=80,
 facecolors="none", zorder=10, edgecolors="k")
plt.scatter(X_train.values[:, 0], X_train.values[:, 1], c=y_train.values, zorder=10, cmap=plt.cm.Paired,
 edgecolors="k")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 8))

sns.scatterplot(x=X_train.values[:, 0], 
                y=X_train.values[:, 1], 
                hue=y_train, 
                s=8);

w = model_C2.coef_[0]          
b = model_C2.intercept_[0]     
x_points = np.linspace(-1, 1)    
y_points = -(w[0] / w[1]) * x_points - b / w[1]  

plt.plot(x_points, y_points, c='r');
```
